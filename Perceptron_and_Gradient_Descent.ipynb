```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

First we need to define a labeled data set. Let's make a toy dataset.

```python
X = np.array([
    [-2, 4],
    [4, 1],
    [1, 6],
    [2, 4],
    [6, 2]
])
```

Next we need to add a bias term -1 into the data set.

```python
X = np.array([
    [-2, 4, -1],
    [4, 1, -1],
    [1, 6, -1],
    [2, 4, -1],
    [6, 2, -1],
])
```

```python
y = np.array([-1, -1, 1, 1, 1])
```

This small toy data set contains two samples labeled with âˆ’1 and three samples labeled with +1. This means we have a binary classification problem, as the data set contains two sample classes. Lets plot the dataset to see, that is is linearly seperable

```python
for d, sample in enumerate(X):
    # Plot the negative samples
    if d < 2:
        plt.scatter(sample[0], sample[1], s=120, marker='_', linewidths=2)
    # Plot the positive samples
    else:
        plt.scatter(sample[0], sample[1], s=120, marker='+', linewidths=2)

# Print a possible hyperplane, that is seperating the two classes.
plt.plot([-2, 6], [6, 0.5])
```

Finally we can code our Perceptron algorithm using our update rule. To keep it simple, we will linearly loop over the sample set. For larger data sets it makes sence, to randomly pick a sample during each iteration in the for-loop.

image.png

The algorithm is shown in the above image.

```python
def perceptron_algo(X, Y):
    w = np.zeros(len(X[0]))
    eta = 1
    epochs = 10

    for epoch in range(epochs):
        for i, x in enumerate(X):
            if (np.dot(X[i], w) * Y[i]) <= 0:
                w = w + eta * X[i] * Y[i]
    return w
```

```python
def perceptron_algo_plot(X, Y):
    '''
    train perceptron and plot the total loss in each epoch.

    :param X: data samples
    :param Y: data labels
    :return: weight vector as a numpy array
    '''
    w = np.zeros(len(X[0]))
    eta = 1
    n = 30
    errors = []

    for t in range(n):
        total_error = 0
        for i, x in enumerate(X):
            if (np.dot(X[i], w) * Y[i]) <= 0:
                total_error += (np.dot(X[i], w) * Y[i])
                w = w + eta * X[i] * Y[i]
        errors.append(total_error * -1)

    plt.plot(errors)
    plt.xlabel('Epoch')
    plt.ylabel('Total Loss')

    return w
```

Next, we will see the the training error at each epoch.

```python
w = perceptron_algo_plot(X, y)
```

We now plot the decision boundary and check which samples have been classified as positive and which samples have been classified as negative.

```python
for d, sample in enumerate(X):
    # Plot the negative samples
    if d < 2:
        plt.scatter(sample[0], sample[1], s=120, marker='_', linewidths=2)
    # Plot the positive samples
    else:
        plt.scatter(sample[0], sample[1], s=120, marker='+', linewidths=2)

# Add our test samples

plt.scatter(2, 2, s=120, marker='_', linewidths=2, color='yellow')
plt.scatter(4, 3, s=120, marker='+', linewidths=2, color='blue')

# Print the hyperplane calculated by perceptron_sgd()
x2 = [w[0], w[1], -w[1], w[0]]
x3 = [w[0], w[1], w[1], -w[0]]

x2x3 = np.array([x2, x3])
X, Y, U, V = zip(*x2x3)
ax = plt.gca()
ax.quiver(X, Y, U, V, scale=1, color='blue')
```

As you can see, our algorithm has correctly classified all the samples. The decision boundary is shown in blue.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons
from sklearn.metrics import accuracy_score

# Perceptron Implementation
def perceptron(X, y, learning_rate=0.01, max_iterations=1000):
    n_samples, n_features = X.shape
    weights = np.zeros(n_features)
    bias = 0
    errors = []

    for epoch in range(max_iterations):
        error_count = 0
        for idx, x_i in enumerate(X):
            linear_output = np.dot(x_i, weights) + bias
            y_predicted = np.where(linear_output > 0, 1, 0)

            update = learning_rate * (y[idx] - y_predicted)
            weights += update * x_i
            bias += update

            if update != 0:
                error_count += 1
        errors.append(error_count)

        # Stop if no errors
        if error_count == 0:
            break

    return weights, bias, errors

# Experiment 1: Toy Dataset (Not Linearly Separable)
X_toy, y_toy = make_moons(n_samples=200, noise=0.2, random_state=42)
y_toy = np.where(y_toy == 0, -1, 1)  # Convert labels to -1 and 1

# Split the toy dataset
X_train, X_test, y_train, y_test = train_test_split(X_toy, y_toy, test_size=0.2, random_state=42)

# Train perceptron on toy dataset
weights, bias, errors = perceptron(X_train, y_train, learning_rate=0.01, max_iterations=1000)

# Plot the error over epochs
plt.figure(figsize=(10, 6))
plt.plot(errors, label="Error Count")
plt.title("Error Count Over Epochs (Toy Dataset)")
plt.xlabel("Epochs")
plt.ylabel("Number of Misclassifications")
plt.legend()
plt.show()

# Test accuracy on toy dataset
y_pred = np.where(np.dot(X_test, weights) + bias > 0, 1, -1)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on Toy Dataset: {accuracy:.2f}")

# Experiment 2: Varying Learning Rate on Toy Dataset
learning_rates = [0.001, 0.01, 0.1, 1.0]

for lr in learning_rates:
    _, _, errors_lr = perceptron(X_train, y_train, learning_rate=lr, max_iterations=1000)
    plt.plot(errors_lr, label=f"LR={lr}")

plt.title("Error Count Over Epochs for Different Learning Rates")
plt.xlabel("Epochs")
plt.ylabel("Number of Misclassifications")
plt.legend()
plt.show()

# Observations
print("\nObservations:")
print("1. With a high learning rate, the algorithm may oscillate and fail to converge.")
print("2. A lower learning rate may require more epochs to converge but can stabilize the learning process.")
print("3. On a non-linearly separable dataset, the perceptron struggles to find a perfect decision boundary.")
```
